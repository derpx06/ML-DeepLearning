🚀 This is what SVM does:

    Takes your data (dots & crosses).

    Tries to find the best line (called a “hyperplane”) that separates them.

    Not just any line — the one that gives the maximum gap (margin) between the two groups.

This is why it's called Support Vector Machine:

    The "support vectors" are the closest points to the separating line from each group.

    These points are super important because they define where the line should go.

    "Machine" because it's an algorithm doing the math.
